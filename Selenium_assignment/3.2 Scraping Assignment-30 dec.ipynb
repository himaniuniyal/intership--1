search results and save it in a data frame and csv. In case if any product has less than 3 pages in search 
results then scrape all the products available under that product name. Details to be scraped are: "Brand 
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and 
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“

```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import time
driver= webdriver.Chrome(r"C:/Users/admin/Downloads/chromedriver_win32/chromedriver.exe")
```

```python
urls=[]
for i in driver.find_elements_by_xpath("//a[@class='a-link-normal s-link-style a-text-normal']"):
    urls.append(i.get_attribute("href"))
```

```python
##the link sometime works some time not, when open in different file. These links shows lots of variation
urls
```

```python
brand=[]
product=[]
price=[]
return_exchange=[]
expected_delivery=[]
availability=[]
url=[]
```

```python
start=0
end=2
for page in range(start,end):
    for i in urls:
        driver.get(i)
        time.sleep(2)
        try:
            brands=driver.find_element_by_xpath("//span[@class='a-size-base']")
            brand.append(brands.text)
        except:
            brand.append('__')
            
        try:
            products=driver.find_element_by_id("productTitle")
            product.append(products.text)
        except:
            product.append('_')
            
        try:
            price1=driver.find_element_by_xpath("//span[@class='a-price-whole']")
            price.append(price1.text)
        except:
            price.append('__')
            
        try:
            ret=driver.find_element_by_xpath("//a[@class='a-size-small a-link-normal a-text-normal']")
            return_exchange.append(ret.text)
        except:
            return_exchange.append('__')
            
        try:
            dele=driver.find_element_by_id("ddmDeliveryMessage")
            expected_delivery.append(dele.text)
        except:
            expected_delivery.append('__')
            
        try:
            avail=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-success']")
            availability.append(avail.text)
        except:
            availability.append('__')
            
        
               
            
```
